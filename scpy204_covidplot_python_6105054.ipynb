```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

อ่านไฟล์

```python
file=pd.read_csv("COVID-19-geographic-disbtribution-worldwide-2020-04-12.csv")
file
```

```python
#Death
ThD=0
CnD=0
VnD=0
AdD=0
ItD=0
LaD=0
TwD=0
GmD=0
BzD=0
JpD=0
```

```python
#Case
ThC=0
CnC=0
VnC=0
AdC=0
ItC=0
LaC=0
TwC=0
GmC=0
BzC=0
JpC=0
```

```python
#เคสสะสม
ATh=[]
ACn=[]
AVn=[]
AAd=[]
AIt=[]
ALa=[]
ATw=[]
AGm=[]
ABz=[]
AJp=[]
```

```python
for j in range(10332):
    i=10331-j
    countries=file.iloc[i]['countriesAndTerritories']
    deaths=file.iloc[i]['deaths']
    cases=file.iloc[i]['cases']
    if countries == 'Thailand' :
        ThD=ThD+deaths
        ThC=ThC+cases
        ATh.append(ThC) 
    if countries == 'Vietnam' :
        VnD=VnD+deaths
        VnC=VnC+cases
        AVn.append(VnC) 
    if countries == 'Andorra' :
        AdD=AdD+deaths
        AdC=AdC+cases
        AAd.append(AdC)
    if countries == 'Italy' :
        ItD=ItD+deaths
        ItC=ItC+cases
        AIt.append(ItC)
    if countries == 'Laos' :
        LaD=LaD+deaths
        LaC=LaC+cases
        ALa.append(LaC)
    if countries == 'Taiwan' :
        TwD=TwD+deaths
        TwC=TwC+cases
        ATw.append(TwC)
    if countries == 'Germany' :
        GmD=GmD+deaths
        GmC=GmC+cases
        AGm.append(GmC)
    if countries == 'Brazil' :
        BzD=BzD+deaths
        BzC=BzC+cases
        ABz.append(BzC)
    if countries == 'Japan' :
        JpD=JpD+deaths
        JpC=JpC+cases
        AJp.append(JpC)
    if countries == 'Canada' :
        CnD=CnD+deaths
        CnC=CnC+cases
        ACn.append(CnC)
```

```python
PercentThDeaths=ThD*100/ThC
PercentThCases=ThC*100/66000000
```

```python
print ('a')
```

```python
print ("Number of case in Thailand:" , ThC ,
       "\n%Case/population:" , PercentThCases ,
       '\n%Death/case:',PercentThDeaths)
```

```python
#พล็อตเส้น
fig=plt.figure()
plt.plot(np.arange(len(ATh))+1,ATh,label='Thailand')
plt.plot(np.arange(len(AVn))+1,AVn,label='Vietnam')
plt.plot(np.arange(len(AAd))+1,AAd,label='Andorra')
plt.plot(np.arange(len(AIt))+1,AIt,label='Italy')
plt.plot(np.arange(len(ALa))+1,ALa,label='Laos')
plt.plot(np.arange(len(ATw))+1,ATw,label='Taiwan')
plt.plot(np.arange(len(AGm))+1,AGm,label='Germany')
plt.plot(np.arange(len(ABz))+1,ABz,label='Brazil')
plt.plot(np.arange(len(AJp))+1,AJp,label='Japan')
plt.plot(np.arange(len(ACn))+1,ACn,label='Canada')

plt.yscale('log')
plt.legend(bbox_to_anchor=(1.02,0.7),loc='upper left')
plt.title("Relation graph between the number of days and number of cases in 10 countries")
plt.xlabel("Number of days")
plt.ylabel("Number of cases")
plt.xlim(0)
```
